```python
import pandas as pd 
import numpy as np 
from sklearn.svm import SVC 

X=np.array([[3,3],[4,3],[1,1]])
Y=np.array([[1],[1],[-1]])
P=np.array([[0.5,1]])
```

```python
'''
（1）用线性可分SVM求【例7-1】的决策函数,依次输出支持向量的总数量、每个类别的支持向量数量，以及支持向量的索引。

（2）索引输出a、W、b的值。

'''
#用线性可分linear进行拟合模型

svc = SVC(C=1, kernel='linear')
svc.fit(X, Y)



#SVC支持向量

# print('SVC共有{}个样本数据，其中{}个支持向量.'.format())
# print('SVC模型支持向量的索引:',)


#决策函数y=wx+b

print('拉格朗日乘子a', svc.dual_coef_)
print('向量W', svc.coef_)
print('常数b', svc.intercept_)
```

```python
svc.support_vectors_
```

```python
'''
（3）利用分类模型的决策函数进行预测新的样本P，输出样本类别及样本到决策边界的距离

'''
#利用决策函数进行预测新样本P1的类别

P = np.array([[0.5, 1]])
svc.decision_function(P)
```
