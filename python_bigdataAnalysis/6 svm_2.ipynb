```python
import pandas as pd 
import numpy as np 
from sklearn.svm import SVC 

X=np.array([[3,3],[4,3],[1,1]])
Y=np.array([[1],[1],[-1]])
P=np.array([[0.5,1]])
```

```python
'''
1SVM7-1,

2aWb

'''
#linear

svc = SVC(C=1, kernel='linear')
svc.fit(X, Y)



#SVC

# print('SVC{}{}.'.format())
# print('SVC:',)


#y=wx+b

print('a', svc.dual_coef_)
print('W', svc.coef_)
print('b', svc.intercept_)
```
